## Data preprocessing

### Subtask:
Identify categorical and numerical features. Apply one-hot encoding to categorical features. (Optional) Apply standard scaling to numerical features. Ensure the training and testing sets are preprocessed consistently.


**Reasoning**:
Identify categorical and numerical columns, apply one-hot encoding to categorical features, apply standard scaling to numerical features, and concatenate the preprocessed features for both training and testing sets.



````python
from sklearn.preprocessing import StandardScaler

# Identify categorical and numerical columns
categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()
# Exclude 'LoanID'
categorical_cols.remove('LoanID')
numerical_cols = X_train.select_dtypes(include=['number']).columns.tolist()

# Apply one-hot encoding to categorical features
X_train_categorical_processed = pd.get_dummies(X_train[categorical_cols], drop_first=True)
X_test_categorical_processed = pd.get_dummies(X_test[categorical_cols], drop_first=True)

# Ensure columns are consistent after one-hot encoding
train_cols = set(X_train_categorical_processed.columns)
test_cols = set(X_test_categorical_processed.columns)

missing_in_test = list(train_cols - test_cols)
for c in missing_in_test:
    X_test_categorical_processed[c] = 0

missing_in_train = list(test_cols - train_cols)
for c in missing_in_train:
    X_train_categorical_processed[c] = 0

X_test_categorical_processed = X_test_categorical_processed[X_train_categorical_processed.columns]


# Apply standard scaling to numerical features
scaler = StandardScaler()
X_train_numerical_scaled = scaler.fit_transform(X_train[numerical_cols])
X_test_numerical_scaled = scaler.transform(X_test[numerical_cols])

# Convert scaled numerical features back to DataFrame
X_train_numerical_scaled_df = pd.DataFrame(X_train_numerical_scaled, columns=numerical_cols, index=X_train.index)
X_test_numerical_scaled_df = pd.DataFrame(X_test_numerical_scaled, columns=numerical_cols, index=X_test.index)

# Concatenate the preprocessed features
X_train_processed = pd.concat([X_train_numerical_scaled_df, X_train_categorical_processed], axis=1)
X_test_processed = pd.concat([X_test_numerical_scaled_df, X_test_categorical_processed], axis=1)

display("Processed X_train head:")
display(X_train_processed.head())
display("Processed X_test head:")
display(X_test_processed.head())
````

## Model selection

### Subtask:
Choose a suitable classification model for predicting loan defaults (e.g., Logistic Regression, Random Forest, Gradient Boosting).


**Reasoning**:
Instantiate Logistic Regression, Random Forest, and Gradient Boosting models with default parameters.



````python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Instantiate Logistic Regression model
logistic_regression_model = LogisticRegression(random_state=42)

# Instantiate Random Forest model
random_forest_model = RandomForestClassifier(random_state=42)

# Instantiate Gradient Boosting model
gradient_boosting_model = GradientBoostingClassifier(random_state=42)

print("Logistic Regression model instantiated:", logistic_regression_model)
print("Random Forest model instantiated:", random_forest_model)
print("Gradient Boosting model instantiated:", gradient_boosting_model)
````

## Model training

### Subtask:
Train the selected classification models using the preprocessed training data.


**Reasoning**:
Train each of the instantiated models using the preprocessed training data.



````python
# Train the Logistic Regression model
logistic_regression_model.fit(X_train_processed, y_train)

# Train the Random Forest model
random_forest_model.fit(X_train_processed, y_train)

# Train the Gradient Boosting model
gradient_boosting_model.fit(X_train_processed, y_train)

print("All models trained successfully.")
````

## Model evaluation

### Subtask:
Evaluate the trained classification models' performance on the preprocessed testing data using relevant metrics such as accuracy, precision, recall, F1-score, and AUC-ROC.


**Reasoning**:
Import the necessary metrics from sklearn.metrics and create a list of the trained models. Then iterate through the models, make predictions, calculate the evaluation metrics, and print the results for each model.



````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# List of trained models
trained_models = {
    "Logistic Regression": logistic_regression_model,
    "Random Forest": random_forest_model,
    "Gradient Boosting": gradient_boosting_model
}

# Evaluate each model
for model_name, model in trained_models.items():
    print(f"Evaluating {model_name}...")

    # Make predictions
    y_pred = model.predict(X_test_processed)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1-score: {f1:.4f}")

    # Calculate AUC-ROC if the model supports predict_proba
    if hasattr(model, "predict_proba"):
        y_pred_proba = model.predict_proba(X_test_processed)[:, 1]
        auc_roc = roc_auc_score(y_test, y_pred_proba)
        print(f"  AUC-ROC: {auc_roc:.4f}")
    else:
        print("  AUC-ROC: Not applicable (model does not support predict_proba)")

    print("-" * 30)
````

## Model saving

### Subtask:
Save the trained classification models to files for future use.


**Reasoning**:
Save the trained models to files using joblib.



````python
import joblib
import os

# Create a directory to save the models if it doesn't exist
if not os.path.exists('trained_models'):
    os.makedirs('trained_models')

# Iterate through the trained_models dictionary and save each model
for model_name, model in trained_models.items():
    # Create a filename from the model name
    filename = f"trained_models/{model_name.lower().replace(' ', '_')}_model.joblib"

    # Save the model to the file
    joblib.dump(model, filename)

    # Print a confirmation message
    print(f"Model '{model_name}' saved to '{filename}'")
````
